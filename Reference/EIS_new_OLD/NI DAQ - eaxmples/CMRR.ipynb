## CMRR

```python
#!/usr/bin/env python3
"""
Simple CMRR Evaluation Script for NI-DAQmx

This script generates a sine sweep on AO0 and measures the differential response
on AI0 (both inputs shorted to AO0) to compute the CommonMode Rejection Ratio (CMRR).

Requirements:
 - nidaqmx
 - numpy
 - matplotlib

Usage:
    python evaluate_cmr_script.py
"""

import numpy as np
import matplotlib.pyplot as plt
import nidaqmx
from nidaqmx.constants import TerminalConfiguration

# === User Parameters ===
device_name   = 'Dev1'                         # Change to your DAQ device
ao_channel    = f"{device_name}/ao0"
ai_channel    = f"{device_name}/ai0"          # AI0 in DIFF mode measures AO0s common mode

# Sweep configuration
frequencies   = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000, 20000, 50000, 60000, 70000]        # Hz
N_periods     = 100                               # Number of sine periods to record at each frequency
amplitude_pk  = 1                             # Peak amplitude in Volts (for 1Vpp use 0.5)
oversample    = 5                              # Samples per period

# Derived sample rate (Hz)
sample_rate = int(max(frequencies) * oversample)

# Storage
measured_pk   = []  # measured peak amplitudes
cmrr_db       = []  # CMRR in dB (20log10(input_pk / measured_pk))

# Loop over frequencies
def measure_at_frequency(freq):
    period           = 1.0 / freq
    samples_per_period = int(sample_rate / freq)
    total_samples      = samples_per_period * N_periods

    # Generate output sine waveform
    t = np.arange(total_samples) / sample_rate
    waveform = amplitude_pk * np.sin(2 * np.pi * freq * t)

    # Create and configure tasks
    with nidaqmx.Task() as ao_task, nidaqmx.Task() as ai_task:
        # AO: finite output
        ao_task.ao_channels.add_ao_voltage_chan(ao_channel)
        ao_task.timing.cfg_samp_clk_timing(sample_rate, samps_per_chan=total_samples)

        # AI: differential input, finite samples
        ai_task.ai_channels.add_ai_voltage_chan(
            ai_channel,
            terminal_config=TerminalConfiguration.DIFF,
            min_val=-1,
            max_val=1

        )
        ai_task.timing.cfg_samp_clk_timing(sample_rate, samps_per_chan=total_samples)

        # Start measurement: AI first, then AO
        ai_task.start()
        ao_task.write(waveform, auto_start=True)

        # Read back the data
        data = np.array(ai_task.read(number_of_samples_per_channel=total_samples))
        ao_task.wait_until_done(timeout=5.0)

    # Compute FFT peak (normalized so sine peak = amplitude_pk)
    fft_vals = np.fft.rfft(data)
    freqs    = np.fft.rfftfreq(total_samples, d=1.0/sample_rate)
    idx      = np.argmin(np.abs(freqs - freq))
    # 2/N normalization gives peak amplitude
    measured = (2.0 / total_samples) * np.abs(fft_vals[idx])
    return measured

for f in frequencies:
    measured = measure_at_frequency(f)
    measured_pk.append(measured)
    # CMRR ratio: input_pk / measured_pk
    cmrr_linear = amplitude_pk / measured if measured != 0 else np.inf
    cmrr_db.append(20.0 * np.log10(cmrr_linear))
    print(f"{f:8.1f} Hz: measured_pk={measured:.3e} V, CMRR={cmrr_db[-1]:.2f} dB")

# Plot Bode of CMRR
plt.figure(figsize=(6,4))
plt.semilogx(frequencies, cmrr_db, marker='o', linestyle='-')
plt.xlabel('Frequency (Hz)')
plt.ylabel('CMRR (dB)')
plt.title('CMRR vs. Frequency')
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
